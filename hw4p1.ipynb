```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
x = [-4.3, 3.2]
mu = np.linspace(-10, 10, num=100)
y = np.exp(-1.0/20*mu**2) * [np.prod([1/(1+(xi-muj)**2) for xi in x]) for muj in mu]
```

```python
px.line(x=mu, y=y, title=r"$\text{Posterior}\  p(\mu \mid x)$", labels={"x": r"$\mu$", "y": "unscaled density"})
```
